```python
import pandas as pd
```

```python
vine_table = pd.read_csv('vine_table.csv')
vine_table.head()
```

```python
votes_df = vine_table['total_votes']
votes_df.head()
```

```python
# Filter total_votes to show only count with greater than 20 votes
total_votes_df = vine_table[vine_table.total_votes >= 20]
votes_df = total_votes_df['total_votes']
total_votes_df.head()
```

```python
helpful_votes_df = total_votes_df['helpful_votes']
helpful_votes_df.head()
```

```python
total_votes_df["helpful / total"] = total_votes_df["helpful_votes"] / total_votes_df["total_votes"]
total_votes_df.head()
```

```python
total_filter_df = total_votes_df[(total_votes_df.df1 > 0.5)]
total_filter_df.head()
```

```python
# Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows where a review was written as part of the Vine program (paid)
vine_df = total_filter_df["vine"]
vine_filter_df = total_filter_df[vine_df == "Y"]
vine_filter_df.head()
```

```python
# Step 3, but N instead of Y
vine_filterN_df = total_filter_df[vine_df == "N"]
vine_filterN_df.head()
```
